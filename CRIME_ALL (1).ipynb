```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load for each crime database
file_to_load_1 = "/Users/fernandawolburg/Downloads/cleancrimedata2016.csv"
file_to_load_2 = "/Users/fernandawolburg/Downloads/cleancrime_2017.csv"
file_to_load_3 = "cleancrime_2018.csv"

# Read the crime_data for each year
crime_2016 = pd.read_csv(file_to_load_1)
crime_2017 = pd.read_csv(file_to_load_2)
crime_2018 = pd.read_csv(file_to_load_3)
crime_2018.head()
```

```python
# Fix the "YEAR" column formatting
crime_2016['YEAR'] = crime_2016['YEAR'].map(lambda x: str(x)[:-2])
crime_2017['YEAR'] = crime_2017['YEAR'].map(lambda x: str(x)[:-2])
crime_2018['YEAR'] = crime_2018['YEAR'].map(lambda x: str(x)[:-2])
```

```python
# DATA AND GRAPH FOR 2016
date_2016 = pd.DataFrame({'date': crime_2016['OCCURRED ON'],
                       'year': crime_2016['YEAR']})
date_2016['TOTAL'] = date_2016.groupby(["date"])["year"].transform('count')
date_2016 = date_2016.drop_duplicates()

date_2016 = date_2016.dropna()

# create a line graph that shows how many crime incidents occurred on each day
fig = plt.figure(figsize=(15,7))
ax = plt.axes()

# create the variables, date on the x axis and total on the y axis and add them to the plot
x = date_2016['date']
y = date_2016['TOTAL']
handle = 'Crime incidents'
handles = 'Full Moon'
ax.plot(x, y,linewidth=1.5, color='g')

# set the limits fo each axis for your graph
plt.xlim(0, 366,) 
plt.ylim(100, 250)

# format the graph to be more visually friendly
plt.title('Crime Counts per-day in 2016')
plt.xlabel('Day')
plt.ylabel('Crime Counts')
plt.xticks([])
plt.axvline(23,linewidth=1, color='y')
plt.axvline(53,linewidth=1, color='y')
plt.axvline(83,linewidth=1, color='y')
plt.axvline(112,linewidth=1, color='y')
plt.axvline(142,linewidth=1, color='y')
plt.axvline(172,linewidth=1, color='y')
plt.axvline(201,linewidth=1, color='y')
plt.axvline(231,linewidth=1, color='y')
plt.axvline(260,linewidth=1, color='y')
plt.axvline(289,linewidth=1, color='y')
plt.axvline(319,linewidth=1, color='y')
plt.axvline(348,linewidth=1, color='y')

#Show the graph
plt.show()
```
